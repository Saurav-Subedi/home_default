```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
pd.set_option('display.max_columns', 100)
sns.set_theme(style="whitegrid")
%matplotlib inline

# Load the feature-engineered training data
df = pd.read_parquet('../data/fe_train_final.parquet')
print(f"Shape: {df.shape}")
df.head()
```

```python
print("=== TARGET Distribution ===")
print(df['TARGET'].value_counts())
print(f"\nDefault rate: {df['TARGET'].mean():.2%}")

fig, ax = plt.subplots(figsize=(6, 4))
df['TARGET'].value_counts().plot(kind='bar', ax=ax, color=['steelblue', 'salmon'])
ax.set_xticklabels(['No Default (0)', 'Default (1)'], rotation=0)
ax.set_title('Class Distribution')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)
missing_df = pd.DataFrame({'missing_count': missing, 'missing_pct': missing_pct})
missing_df = missing_df[missing_df['missing_count'] > 0].sort_values('missing_pct', ascending=False)

print(f"Columns with missing values: {len(missing_df)} / {df.shape[1]}")
print(missing_df.head(20))
```

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 4))

df['AMT_INCOME_TOTAL'].clip(upper=df['AMT_INCOME_TOTAL'].quantile(0.99)).hist(
    bins=50, ax=axes[0], color='steelblue')
axes[0].set_title('Income Distribution')

df['AMT_CREDIT'].hist(bins=50, ax=axes[1], color='steelblue')
axes[1].set_title('Credit Amount')

df['AMT_GOODS_PRICE'].hist(bins=50, ax=axes[2], color='steelblue')
axes[2].set_title('Goods Price')

plt.tight_layout()
plt.show()
```

```python
numeric_df = df.select_dtypes(include=[np.number])
corr = numeric_df.corr()['TARGET'].drop('TARGET').abs().sort_values(ascending=False)

print("=== Top 20 features correlated with TARGET ===")
print(corr.head(20))

corr.head(20).plot(kind='bar', figsize=(14, 5), color='steelblue')
plt.title('Top 20 Feature Correlations with TARGET')
plt.tight_layout()
plt.show()
```

## EDA Key Findings

- **Default rate: 8.07%** — dataset is imbalanced (ratio ~11:1). Use `scale_pos_weight=11` in LightGBM.
- **147/202 features have missing values** — top missing: BUREAU_AMT_ANNUITY (74%), COMMONAREA (70%), OWN_CAR_AGE (66%). LightGBM handles NaN natively — no imputation needed.
- **Top predictors**: EXT_SOURCE_3, EXT_SOURCE_2, EXT_SOURCE_1 (external credit scores) — strongest signal by far.
- **Income distribution**: right-skewed, clipped at 99th percentile for visualization.
